# Web Scraping data Jumlah penduduk menurut Provinsi

sumber : https://id.wikipedia.org/wiki/Demografi_Indonesia

```python
#import library  urllib
from urllib.request import urlopen
html = urlopen("https://id.wikipedia.org/wiki/Demografi_Indonesia").read()
print(html[0:1000])
```

```python
#import BeautifulSoup
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, "lxml")
print(soup.prettify()[0:1000])
```

```python
#buat list-list kosong
kode_bps = []
nama = []
ibu_kota = []
populasi = []
luas = []
pulau = []

#measukan data ke dalam list
data = soup.find("table", {'class': "wikitable sortable"}).find_all("tr")[1:]
for i in data:
    daftar = i.find_all("td")
    kode_bps.append(daftar[0].get_text())
    nama.append(daftar[2].find("a").get_text())
    ibu_kota.append(daftar[4].get_text())
    populasi.append(int(daftar[5].get_text().replace(".","")))
    luas.append(float(daftar[6].get_text().replace(".","").replace(",",".")))
    pulau.append(daftar[8].get_text().replace("\n",""))
```

```python
#menggabungakan data dalam sebuah dictonary
jumlah_populasi = {
    'Kode BPS' : kode_bps,
    'Nama' : nama,
    'Ibu Kota' : ibu_kota,
    'Populasi' : populasi,
    'Luas km' : luas,
    'Pulau' : pulau,
}
```

```python
#import pandas
import pandas as pd
df = pd.DataFrame(jumlah_populasi, columns=list(jumlah_populasi.keys()))
df
```

```python
# 5 Provinsi terpadat di Indonesia
df.sort_values("Populasi", ascending = False, ignore_index = True).head(5)
```

```python
# jumlah popoulasi berdasarkan pulau
df.groupby(['Pulau']).sum().sort_values('Populasi', ascending = False)
```

```python
#import ke dalam file CSV
df.to_csv("Jumlah_populasi_Indonesia.csv", index=False,encoding='utf', quoting = 1)
```
